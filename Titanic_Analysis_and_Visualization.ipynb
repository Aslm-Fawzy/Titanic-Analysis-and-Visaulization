# **Titanic Analysis and Visualization**

Of the estimated 2,224 passengers and crew aboard, more than 1,500 died, making the sinking at the time one of the deadliest of a single ship[a] and the deadliest peacetime sinking of a superliner or cruise ship to date.

#**Importing Libraries and Dataset**





```python
import numpy as np 
import pandas as pd 
import seaborn as sns
sns.set(rc={'figure.figsize': [12, 8]}, font_scale=1.3)
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df_titanic = sns.load_dataset('titanic')
```

# **EDA**

```python
df_titanic.fillna(df_titanic.mean(), inplace=True)
df_titanic.info()
```

```python
df_titanic[df_titanic.embark_town.isnull()]
```

```python
df_titanic.drop(index=[61,829],inplace=True)
df_titanic.drop(columns = 'deck',axis=1,inplace=True)
df_titanic.info()
```

## **Classes Of  Passenger in Titanic**

```python
sns.set(rc={'figure.figsize': [12, 8]}, font_scale=1.3)
sns.countplot(x='class',data=df_titanic,palette='mako')
```

##**Number of passenger in each embarked town** 

```python
sns.countplot(x='embark_town',data=df_titanic,palette='mako')
```

##**Human categories In Titanic**   

```python
sns.countplot(x='who',data=df_titanic,palette='mako')
```

##**Ditribution of Passenger fares**

```python
sns.kdeplot(x='fare',data=df_titanic,color = 'k',shade=True,alpha = .5)
```

```python
sns.displot(x='fare',data=df_titanic,palette='mako',bins=50,height=8,aspect=1.2)
```

## **Average of fares for Survived and Not Survived**

```python
sns.barplot(x=df_titanic['alive'], y=df_titanic['fare'],palette='mako')
```

##**Average of fares in each embarked town**

```python
sns.barplot(x='embark_town',y='fare',data=df_titanic,palette='mako')
```

##**Average of fares in each class**

```python
sns.barplot(x='class',y='fare',data=df_titanic,palette='mako')
```

##**Ditribution of Passenger age**

```python
sns.kdeplot(x='age',data=df_titanic,color = 'k',shade=True,alpha= .5)
```

```python
sns.displot(x='age',data=df_titanic,palette='mako',bins=30,height=8,aspect=1.2 )
```

As we see from the graphs, Most of passengers age are in range 25-35 years 

## **Survived Vs Not Survived in Titanic Diseater**

## **The Survived and not Survived Categories in Titanic Diseater**

```python
sns.countplot(x='alive',data=df_titanic,hue='who')
```

```python
sns.barplot(x='who',y='survived',data=df_titanic,estimator = np.sum)
```

As we see from the graphs , Most of survived category  in titanic is woman.

## **Average of Age of Survived and Not Survived**

```python
sns.kdeplot(x='age',data=df_titanic,shade=True,hue = df_titanic['alive'],multiple = 'stack')
```

```python
sns.barplot(x='alive',y='age',data=df_titanic)
```

average of age for survived and not survived is approximatly equal

## **If passenger is alone is it chance to be survived increase ?** 

```python
sns.barplot(x='alone',y='survived', data=df_titanic,palette='mako')
```

```python
df_titanic['passenger companions']=df_titanic['sibsp']+df_titanic['parch']
```

##**Relation between age , survived , class , fare and  passenger companions**
